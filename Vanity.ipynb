```python
#get last 4/5/6 digits from TFN
df['lastFour'] = df['TFN'].apply(lambda x: str(x)[-4:])
df['lastFive'] = df['TFN'].apply(lambda x: str(x)[-5:])
df['lastSix'] = df['TFN'].apply(lambda x: str(x)[-6:])
```

```python
#maps number to dialpad and gives all possible combination

letters_map = {'2':'ABC', '3':'DEF', '4':'GHI', '5':'JKL', 
               '6':'MNO', '7':'PQRS', '8':'TUV', '9':'WXYZ'}

def possible_words(phone_number):
    phone_number = str(phone_number)
    try:
        letters_to_combine = (letters_map[digit] for digit in phone_number)
        for letters_group in itertools.product(*letters_to_combine):
            yield ''.join(letters_group)
    except:
        return ''
```

```python
#English dictionaries witth all 4/5/6 letter words

all_words = pd.DataFrame(words.words())
all_words4 = all_words[all_words[0].apply(lambda x: len(str(x)))==4]
all_words5 = all_words[all_words[0].apply(lambda x: len(str(x)))==5]
all_words6 = all_words[all_words[0].apply(lambda x: len(str(x)))==6]
all_words7 = all_words[all_words[0].apply(lambda x: len(str(x)))==7]
all_words8 = all_words[all_words[0].apply(lambda x: len(str(x)))==8]
```

```python
#Functions that check in dictionaries
def eng_words(phone_number):
    'converts number into letter combinations and checks in respective dictionaries'
    x=[]
    for i in list(possible_words(phone_number)):
        if i.lower() in words.words():
            x.append(i)
    return x


def eng_words4(phone_number):
    'Check English dict with all 4 letter'
    x=[]
    for i in list(possible_words(phone_number)):
        if i.lower() in list(all_words4[0]):
            x.append(i)
    return x


def eng_words5(phone_number):
    'Check English dict with all 5 letter'
    x=[]
    for i in list(possible_words(phone_number)):
        if i.lower() in list(all_words5[0]):
            x.append(i)
    return x


def eng_words6(phone_number):
    'Check English dict with all 6 letter'
    x=[]
    for i in list(possible_words(phone_number)):
        if i.lower() in list(all_words6[0]):
            x.append(i)
    return x

def eng_words7(phone_number):
    'Check English dict with all 7 letter'
    x=[]
    for i in list(possible_words(phone_number)):
        if i.lower() in list(all_words7[0]):
            x.append(i)
    return x

def eng_words8(phone_number):
    'Check English dict with all 8 letter'
    x=[]
    for i in list(possible_words(phone_number)):
        if i.lower() in list(all_words8[0]):
            x.append(i)
    return x
```
